## Homing Dobot 

It can happen, that the dobot is not in the desired home position at the beginning of the program. In this case, you can use the following code to program the desired home position for the dobot. 

In a first step you need to execute the code below. With that you will place the dobot at the current homing position. In addition all available interface methods of the dobot class will be shown.

```python
from time import sleep
from dobotmaster.lib.dobot import Dobot
from dobot_functions import find_dobot_ports

ports = find_dobot_ports()

for i, port in enumerate(ports):
    print(f"\n=== Dobot {i} at port {port} ===")
    bot = Dobot(port)
    bot.interface.clear_alarms()
    bot.home()
    bot.interface.wait_until_done()
    sleep(2)
    pose = bot.get_pose()
    print(f"Pose after homing: {pose}")
    print("Available interface-methods:")
    print([m for m in dir(bot.interface) if 'home' in m.lower() or 'param' in m.lower()])
    bot.interface.close_col()
```

Now you need to take a look at the saved home position of the dobot. 

```python
for port in ports:
    print(f"\n--- {port} ---")
    bot = Dobot(port)
    print("Homing-Params before:", bot.interface.get_homing_paramaters())
    print("End-Effector-Params:", bot.interface.get_end_effector_params())
    bot.interface.close_col()
```

You can now change the coordinates for the homing in the code below and execute it. The dobot will move to the new home position and save it there.

! Since this code is designed for a set-up with multiple dobots, you may need to change the port of the dobot in the first line of this code. 

```python
bot = Dobot('/dev/ttyACM1') # change it to the port of your dobot

bot.interface.set_homing_parameters(209.70, 0.0, 100.0, 0.0)
print("Homing parameters set. Start homing...")

# Refresh the home so that values take effect
bot.home()
bot.interface.wait_until_done()
sleep(2)

# Verify the new homing position
print("New pose after homing:", bot.get_pose())
print("New homing-params:", bot.interface.get_homing_paramaters())

bot.interface.close_col()
```
