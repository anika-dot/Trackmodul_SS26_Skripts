## Define Position of Dobot

To define the position of the dobot, you can use the following code. You can change the values of x, y and z to define the desired position. 

The safe_move function is used, that the Dobot does not crash into something if he moves. With that function the dobot first moves to a safe height, then moves to the desired x and y coordinates and finally moves down to the desired z coordinate. 

It's important to home the dobot first with init_and_home_dobot, so that the dobot knows where it is and can move to the desired position. 

```python
from dobot_functions import find_dobot_ports, init_and_home_dobot, safe_move
import time

ports = find_dobot_ports()
print(ports)

# Richtige Ports wählen!
dobot = init_and_home_dobot(ports[1])

time.sleep(1)

print("dobot:", dobot)

# Positionen
# Home: 209.6999969482422, 0.0, 100.0, 0.0
pick_color_sensor_sorter = (150, -190, 50, 65)
throw_position_sorter = (250, 100, 50, 0) 
conveyor_position_sorter = (230, -50, 50, 0)
pick_conveyor_belt_pick_place = (230, 85, 30, 55)
place_color_sensor_pick_place = (150, 255, 50, 45)

safe_move(dobot, place_color_sensor_pick_place)

time.sleep(0.5)

# ERST AM ENDE schließen
dobot.interface.close_col()
```
